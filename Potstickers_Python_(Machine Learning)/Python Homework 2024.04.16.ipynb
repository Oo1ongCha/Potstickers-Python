```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dir_data = 'D:\\Google Drive\\Pending\\Potstickers Python\\Python Homework 2024.04.16'

years = []
approved_funds = []
max_funds_uni = []
min_funds_uni = []
top10_funds_uni = []
bot10_funds_uni = []

for i in range(107,112):
    locals()['csv_'+str(i)] = os.path.join(dir_data, str(i)+'.csv')
    locals()['df_'+str(i)] = pd.read_csv(locals()['csv_'+str(i)],encoding='big5')
    
    #print(str(i)+'å¹´')
    #print(locals()['df_'+str(i)])
    #print(locals()['df_'+str(i)].info())
    
    if locals()['df_'+str(i)].iloc[0:,3].dtypes == 'object':
        locals()['df_'+str(i)].iloc[0:,3] = locals()['df_'+str(i)].iloc[0:,3].str.replace(',','')
        locals()['df_'+str(i)].iloc[0:,3] = locals()['df_'+str(i)].iloc[0:,3].str.replace('-','0')
        locals()['df_'+str(i)].iloc[0:,3] = pd.to_numeric(locals()['df_'+str(i)].iloc[0:,3])
    
    n = locals()['df_'+str(i)].iloc[0:,3]
    m = n.sum()
    
    approved_funds.append(m)
    years.append(i)
    
    # Maximum funding
    max_funds = n.max()
    max_funds_uni.append(locals()['df_'+str(i)].iloc[0:,2][locals()['df_'+str(i)].iloc[0:,3] == max_funds])
    
    # Top 10 funding
    #top10_funds = n.nlargest(10, keep='all')
    top10_funds = n.sort_values()[-10:][::-1]
    tfu=[]
    for j in top10_funds:
        tfu.append(locals()['df_'+str(i)].iloc[0:,2][locals()['df_'+str(i)].iloc[0:,3] == j])
    top10_funds_uni.append(tfu)
    del tfu

    # Minimum funding
    min_funds = n.min()
    min_funds_uni.append(locals()['df_'+str(i)].iloc[:,2][locals()['df_'+str(i)].iloc[0:,3] == min_funds])
    
    # Bottom 10 funding
    #bot10_funds = n.nsmallest(10, keep='all')
    bot10_funds = n.sort_values()[:10]
    bfu=[]
    for j in bot10_funds:
        bfu.append(locals()['df_'+str(i)].iloc[:,2][locals()['df_'+str(i)].iloc[0:,3] == j])
    bot10_funds_uni.append(bfu)
    del bfu

del i, j, n, m, max_funds, top10_funds, min_funds, bot10_funds

plt.scatter(years, approved_funds)

coef = np.polyfit(years, approved_funds,1)
reg_model = np.poly1d(coef)
print ('The Regression Model of funding from 107 to 111 is ',reg_model)
print()

plt.plot(years, reg_model(years), color='red')
plt.xticks(years)

plt.show()
print()

#Maximum funding in year
for i in range (5) :
    print ( 'Maximum funding in year', years[i], 'is', max_funds_uni[i].values)
print()

#Minimum funding in year
for i in range (5) :
    print ( 'Minimum funding in year', years[i], 'is', min_funds_uni[i].values)
print()

#Top 10 funding in year
print ('Top 10 funding in year')

top10_funds_uni_df = pd.DataFrame(columns = years, index = list(range(1,11)))
for i in range (5):
    for j in range (10):
        top10_funds_uni_df.iloc[j,i] = top10_funds_uni[i][j].values

print (top10_funds_uni_df)
print()

#Bottom 10 funding in year
print ( 'Bottom 10 funding in year')

bot10_funds_uni_df = pd.DataFrame(columns = years, index = list(range(1,11)))
for i in range (5):
    for j in range (10):
        bot10_funds_uni_df.iloc[j,i] = bot10_funds_uni[i][j].values

print (bot10_funds_uni_df)
```
